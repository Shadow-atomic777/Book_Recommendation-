```python
#importing all libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
#importing kaggle dataset
books=pd.read_csv("../data/raw/books.csv")
ratings=pd.read_csv("../data/raw/ratings.csv")
print(books.head())
print(ratings.head())
```

```python
books.shape,ratings.shape
```

```python
books.isnull().sum()
```

```python
ratings["rating"].describe()
```

```python
book_ratings=ratings.groupby('book_id').agg(
    avg_rating=('rating','mean'),
    num_of_ratings=('rating','count')).reset_index()
book_ratings.head()
```

```python
#merging both datasets
book_ratings=book_ratings.merge(books,on="book_id")
book_ratings.head()
```

```python
#rarely rated books
rarely_rated_books=book_ratings[book_ratings['num_of_ratings']>=50]
rarely_rated_books.tail()
```

```python
#Sorting the books on the basis of average ratings
Top_rated_books=rarely_rated_books.sort_values(by="avg_rating",ascending=False).head(20)
Top_rated_books.head()
```

```python
Top_rated_books=rarely_rated_books.assign(
    score=rarely_rated_books["avg_rating"]*(rarely_rated_books["num_of_ratings"]/
    (rarely_rated_books["num_of_ratings"]).max())
).sort_values(by="score",ascending=False).head(20)  
Top_rated_books.head()
columns=Top_rated_books.columns
print(columns)
```
